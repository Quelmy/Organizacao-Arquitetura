Slide 28 - Faça um programa que o usuário entre com o número de níveis de quantização e receba como resposta o número de bits e todas as combinações desses bits.

```python
# Solicita ao usuário o número de níveis de quantização
niveis = int(input("Digite o número de níveis de quantização: "))

# Inicializa a quantidade de bits com 0
bits = 0

# Enquanto 2 elevado a 'bits' for menor que o número de níveis,
# continua incrementando a quantidade de bits
# Isso garante que teremos bits suficientes para representar todos os níveis
while (2 ** bits) < niveis:
    bits += 1

# Exibe a quantidade mínima de bits necessários
print("\nNúmero de bits necessários:", bits)

# Exibe todas as combinações possíveis com essa quantidade de bits
print("\nCombinações possíveis:")

# Gera todas as combinações possíveis de 0 até 2^bits - 1
for i in range(2**bits):

    # Converte o número para binário com preenchimento de zeros à esquerda
    # Ex: se bits = 3 → 001, 010, 011...
    combinacao = format(i, f'0{bits}b')

    # Mostra a combinação
    print(combinacao)
```

Slide 52 Jupyter: faça um programa que converta de decimal para binário. E para ser explicado em sala de aula

```python
# Entrada do usuário
decimal = int(input("Digite um número decimal: "))

# Variável para armazenar o resultado
binario = ""

# Caso especial: número 0
if decimal == 0:
    binario = "0"
else:
    # Processo de divisão sucessiva por 2
    while decimal > 0:
        resto = decimal % 2
        binario = str(resto) + binario  # adiciona na frente
        decimal = decimal // 2

# Saída
print("Número em binário:", binario)
```

Slide 59 Jupyter: faça um programa que converta um número fracionário decimal para pinário, e vice-versa. Explicar em sala de aula.

```python
# Entrada do usuário
numero = float(input("Digite um número decimal (ex: 10.25): "))

# Separando parte inteira e fracionária
parte_inteira = int(numero)
parte_fracionaria = numero - parte_inteira

bin_inteiro = ""

if parte_inteira == 0:
    bin_inteiro = "0"
else:
    while parte_inteira > 0:
        bin_inteiro = str(parte_inteira % 2) + bin_inteiro
        parte_inteira //= 2

bin_fracionario = ""
contador = 0

# Limite de 10 casas para não entrar em loop infinito
while parte_fracionaria > 0 and contador < 10:
    parte_fracionaria *= 2
    bit = int(parte_fracionaria)
    bin_fracionario += str(bit)
    parte_fracionaria -= bit
    contador += 1

binario = bin_inteiro + "." + bin_fracionario

print("Decimal → Binário:", binario)


binario_input = input("\nDigite um binário (ex: 101.01): ")

if "." in binario_input:
    parte_int, parte_frac = binario_input.split(".")
else:
    parte_int, parte_frac = binario_input, ""

decimal = 0

# Parte inteira
for i, bit in enumerate(parte_int[::-1]):
    decimal += int(bit) * (2**i)

# Parte fracionária
for i, bit in enumerate(parte_frac):
    decimal += int(bit) * (2**-(i+1))

print("Binário → Decimal:", decimal)
```

Slide 75 Jupyter: faça um programa que converta de octal e Hexa para binário e vice-versa. Usando esse programa converta também para decimal usando o binário como intermediário. Explicar em sala de aula.

```python
# Entrada do usuário
numero = input("Digite um número: ")
base = int(input("Base (8 para octal, 16 para hexadecimal): "))

decimal = int(numero, base)

binario = bin(decimal)[2:]

print("\nBinário:", binario)
print("Decimal:", decimal)


octal = oct(decimal)[2:]
hexa = hex(decimal)[2:]

print("Octal:", octal)
print("Hexadecimal:", hexa)
```
